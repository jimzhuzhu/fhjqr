# 随机绘图
 - clone() 创建并返回具有相同位置等属性的乌龟。
 - 随机数：获得随机的数据

```python
# 随机数案例 绘画爱心

import turtle as t
import random as r
def pink():
    color = (1,r.random(),1)
    return color
def heart(r,a):
    # 改变海龟的方向
    t.seth(a)
    t.circle(-r,180)
    t.fd(2 * r)
    t.right(90)
    t.fd(2 * r)
    t.circle(-r,180)
# 设置画布
t.setup(800,800,200,200)
t.speed(0)
t.penup()
t.bgcolor("green")
for i in range(30):
    t.setx(r.randint(-400,400))
    t.sety(r.randint(-400,400))
    t.begin_fill()
    t.fillcolor(pink())
    heart(r.randint(10,50),r.randint(0,90))
    t.end_fill()
```

```python
# 加入clone（） 模拟星空程序

from turtle import *
from random import random,randint
screen = Screen()
width,height = 800,600
screen.setup(width,height)
screen.title("模拟星空_海龟画图")
screen.bgcolor("black")
# 这里要设为0，否则会很卡
screen.delay(0)
t = Turtle(visible = False,shape = "circle")
t.pencolor("white")
t.fillcolor("white")
t.penup()
t.setheading(-90)

stars = []
for i in range(200):
    star = t.clone()
    s = random()/3
    print(s)
    star.shapesize(s,s)
    star.speed(int(s * 10))
    star.setx(width/2 + randint(1,width))
    star.sety(randint(-height/2, height/2))
    star.showturtle()
    stars.append(star)
# 让星空一直显示
while True:
    for star in stars:
        star.setx(star.xcor() - 3 * star.speed())

    if star.xcor() < -width/2:
        star.hideturtle()
        star.setx(width/2 + randint(1,width))
        star.sety(randint(-height/2, height/2))
        star.showturtle()
while True:
    for star in stars:
        star.setx(star.xcor() - 3 * star.speed())
```

# 鼠标处理
 - 绑定参数，调用函数
 - 默认1为鼠标左键
 - 2 为鼠标滑轮
 - 3 为鼠标右键
 - 绑定方法 
     - turtle.onclick(turn,2)
     - turn.onrelease(turn) 弹起

```python
# turtle 鼠标绑定案例
import turtle 
def turn(x,y):
    turtle.lt(90)
    # 打印当前坐标
    print("x=", x, "y=", y)
    # 打印当前方向
    print(turtle.heading())
    # 清除绑定关系
    turtle.onclick(None)
    # 再次绑定
    turtle.onclick(turn)
# 绑定turn方法
turtle.onclick(turn,2)
```

# 递归树
 - 递归作为一种算法在程序设计语言中使用广泛
 - 递归结合clone()

```python
# 递归树

import turtle

def draw_branch(branch_length):
    # 限定绘制的树枝（包括树干、树枝和树叶）长度至少大于5
    if branch_length > 5:
        # 如果长度小于20，即可判定是树叶，绘制成绿色
        if branch_length <= 20:
            turtle.color("green")
        else:
            turtle.color("brown")
        # 绘制树干
        turtle.forward(branch_length)
        turtle.right(20)
        # 绘制每个节点分叉右侧的树枝（递归）
        draw_branch(branch_length - 15)
        
        turtle.left(40)
        # 绘制每个节点分叉左侧的树枝
        draw_branch(branch_length - 15)
        # 如果长度大于20，即可判定是树干，绘制成棕色
        if (branch_length > 20):
            turtle.color("brown")
        turtle.right(20)
        # 左右侧树枝绘制好，返回节点处
        turtle.backward(branch_length)
def main():
    turtle.left(90)
    turtle.penup()
    turtle.backward(150)
    turtle.pendown()
    turtle.speed(0)
    turtle.color("brown")
    # 树枝初始长度为125，长度依次减15
    draw_branch(125)
    
main()
```

```python
# 丰富递归树
from turtle import *
from random import *
from math import *
def tree(n, l):
    pd() # 下笔
    # 阴影效果
    t = cos(radians(heading() + 45)) / 8 + 0.25
    pencolor(t, t, t)
    pensize(n / 3)
    forward(l) # 画树枝
    
    if n > 0:
        b = random() * 15 + 10 # 右分支偏转角度
        c = random() * 15 + 10 # 左分支偏转角度
        d = l * (random() * 0.25 + 0.7) # 下一个分支的长度 # 右转一定角度，画右分支
        right(b)
        tree(n - 1, d)
        # 左转一定角度，画左分支
        left(b + c)
        tree(n - 1, d) # 转回来
        right(c)
    else:
        # 画叶子
        right(90)
        n = cos(radians(heading() - 45)) / 4 + 0.5
        pencolor(n, n*0.8, n*0.8)
        circle(3)
        left(90)
        
        # 添加0.3倍的飘落叶子
        if (random() > 0.7):
            pu()
            # 飘落
            t = heading()
            an = -40 + random() * 40
            setheading(an)
            dis = int(800*random()*0.5 + 400*random()*0.3 + 200*random()*0.2)
            forward(dis)
            setheading(t)
            # 画叶子 pd()
            right(90)
            n = cos(radians(heading() - 45)) / 4 + 0.5
            pencolor(n*0.5 + 0.5, 0.4 + n*0.4, 0.4 + n*0.4)
            circle(2)
            left(90)
            pu()
            # 返回
            t = heading()
            setheading(an)
            backward(dis)
            setheading(t)
    pu()
    backward(l) # 返回
bgcolor(0.5, 0.5, 0.5) # 背景色
ht() # 隐藏turtle
speed(0) # 速度，1,10渐快，0最快
tracer(0, 0)
pu() # 抬笔
backward(100)
left(90) # 左转90度
pu() # 抬笔
backward(300) # 后退300
tree(12, 100) # 递归7层
done()
```

# 代码动画
 - 绝对坐标
 - 视觉暂留效应
 - 动画原理
 - goto语句不需要频繁的转向

```python
# 绘画小鱼

import turtle as t
t.speed(1)

# 身体
t.goto(100,100)
t.goto(100,-100)
t.goto(0,0)
# 头
t.goto(50,50)
t.goto(50,-50)
t.goto(0,0)
# 眼睛
t.up()
t.goto(30,0)
t.left(90)
t.down()
t.circle(5)
# 尾巴
t.up()
t.goto(100,0)
t.down()
t.goto(150,50)
t.goto(150,-50)
t.goto(100,0)
```

```python
# 让绘画小鱼动起来
import turtle as t
import time
def fish(i):
    # 身体
    t.goto(100-i,100)
    t.goto(100-i,-100)
    t.goto(0-i,0)
    # 头
    t.goto(50-i,50)
    t.goto(50-i,-50)
    t.goto(0-i,0)
    # 眼睛
    t.up()
    t.goto(30-i,0)
    t.left(90)
    t.down()
    t.circle(5)
    t.right(90)
    # 尾巴
    t.up()
    t.goto(100-i,0)
    t.down()
    t.goto(150-i,50)
    t.goto(150-i,-50)
    t.goto(100-i,0)
    t.up()
    t.goto(0-i,0)
    t.down()
for i in range(1,200):
    # 画画的时间
    t.delay(0.1)
    fish(i)
    t.clear()
```

```python
# 让绘画小鱼动起来
# Shape专用作图方式，参数固定式compound

import turtle as t
# t.hideturtle()
# 默认添加一个组合画图
s = t.Shape("compound")

body = ((0,0), (100,100), (-100,100))
head = ((0,0), (50,50), (-50,50))
eye = ((5,30), (5,35), (-5,30), (-5,35))
tail = ((0,100), (50,150), (-50,150))

s.addcomponent(body, "yellow", "red")
s.addcomponent(head, "blue", "red")
s.addcomponent(eye, "black", "red")
s.addcomponent(tail, "yellow", "red")

t.register_shape("myshape", s)
t.shape("myshape")
t.up()
t.speed(0)
for i in range(1,200):
    t.backward(10)
t.down()
```
